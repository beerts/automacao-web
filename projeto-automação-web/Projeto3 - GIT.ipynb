# Automação Web e Busca de Informações com Python
Desafio:
Trabalhamos em uma importadora e compramos e vendemos commodities:

Soja, Milho, Trigo, Petróleo, etc.
Precisamos pegar na internet, de forma automática, a cotação de todas as commodites e ver se ela está abaixo do nosso preço ideal de compra. Se tiver, precisamos marcar como uma ação de compra para a equipe de operações.

Base de Dados: https://drive.google.com/drive/folders/1O9aE_Hen5smf_a6TsbVF6DM0ms4Ykj6s?usp=share_link

Para isso, vamos criar uma automação web:

Usaremos o selenium
Importante: baixar o webdriver

```python
#1- ENTRAR NA INTERNET(abrir navegador)
from selenium import webdriver as wd
navegador = wd.Chrome()#2- IMPORTAR BASE DE DADOS
navegador.get('https://www.google.com')


#2- IMPORTAR BASE DE DADOS
import pandas as pd

tabela = pd.read_excel("commodities.xlsx")
display(tabela)

#.send_keys("  ") -> escrever nele
#.click() -> clicar nele
#.get_attribute() -> pegar uma informação dele

for linha in tabela.index:
    
    produto = tabela.loc[linha, "Produto"]       #forma mais pratica sem ter que ficar utilizando o replace
                                                 #import unicodedata
                                                 #produto = unicodedata.normalize("NFKD", produto).encode("ascii", "ignote")
    produto = produto.replace("á", "a").replace("ó", "o").replace("ã", "a").replace("ç","c").replace("ú","u").replace("é", "e")
    
    link = f"https://www.melhorcambio.com/{produto}-hoje"
    print(link)
    navegador.get(link)
    

    #pegar cotação do milho
    preco = navegador.find_element('xpath', '//*[@id="comercial"]').get_attribute('value')
    preco = preco.replace(".", "").replace(",", ".")
    print(preco)
    tabela.loc[linha, "Preço Atual"] = float(preco)


#3- PARA CADA PRODUTO NA NOSSA BASE DE DADOS
#4- PEGAR PREÇO ATUAL DO PRODUTO
#5- ATUALIZAR O PREÇO NA BASE DE DADOS
#6- DECIDIR QUAIS PRODUTOS A GENTE VAI COMPRAR

#preencher a coluna comprar
tabela["Comprar"]= tabela["Preço Atual"] < tabela["Preço Ideal"]
display(tabela)

#7- EXPORTAR A BASE DE DADOS ATUALIZADA
tabela.to_excel("commodities_atualizado.xlsx", index=False)

    

```
